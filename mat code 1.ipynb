```python
import matplotlib.pyplot as plt
# Define the points
x = [4, 0]
y = [0, 6]
# Create the plot
plt.plot(x, y, 'o-')
# Set the axis limits
plt.xlim(0, 5)
plt.ylim(0, 7)
# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph Joining Points (4, 0) and (0, 6)')
# Display the plot
plt.show()
```
